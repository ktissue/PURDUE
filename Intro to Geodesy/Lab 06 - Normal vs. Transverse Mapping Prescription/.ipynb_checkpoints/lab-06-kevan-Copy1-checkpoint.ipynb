**Function for converting DMS to Decimal Degrees**

```python
def dms2dd(dd, mm, ss):
    """
    A function that will convert degree minute second (DMS) format into decimal degree format
    
    Inputs
    -----
    dd: spherical coordinate value: degrees (integer)
    mm:spherical coordinate value: minutes (integer)
    ss: spherical coordinate value: seconds (floating number)
    
    Output
    ------
    calculated decimal degree value (floating number)
    """
    return dd + mm/60 + ss/3600
```

## 1.3: Map: Mercator

```python
coords_dd = pd.DataFrame(
    {"psi":[psi_dd],
     "lambda":[lmd_dd]})
```

**Mercator Mapping Prescription**

```python
def mercator(df):
    """
    A function that will take a data frame containing coordinates in decimal degrees
    and apply a mercator mapping prescription to it
    
    Input
    -----
    pandas.DataFrame with columns: psi, lambda (no prescription)
    
    Output
    ------
    pandas.DataFrame with columns: psi, lambda (mercator prescription)
    """
    
    # limiting the latitude to +/- 75 degrees (before mercator)
    df = df[(-80 <= df.psi) & (df.psi <= 81)]
    
    # resetting the data frame index
    df = df.reset_index(drop=True)
    
    # applying the mercator mapping prescription 
    df.psi = np.log(np.tan(((df.psi*rad)/2) + (math.pi/4))) * deg
    
    # limiting the latitude to +/- 75 degrees again (after mercator)
    df = df[(-80 <= df.psi) & (df.psi <= 81)]
    
    # resetting the data frame index
    df = df.reset_index(drop=True)

    return df
```

**Plotting the Mercator Mapping Prescription**

```python
# Applying the Mercator Mapping Prescription to the coordinates
coords_M = mercator(coords_dd)

Xm_M = coords_M['lambda']
Ym_M = coords_M['psi']
```

```python
# Labels for markers
labels_M = ['  PoB']

# Resizing the plot window
plt.rcParams["figure.figsize"] = (14,7)

# Plotting the original locations
plt.plot(Xm_M, Ym_M, 'ro', markersize=10, zorder=5) # SF

# Applying the Mercator Mapping Prescription to the coastlines
coast_M = mercator(coast_dd)

# Plotting the coastlines
plt.scatter(coast_M['lambda'], coast_M['psi'], c='k', s=4, edgecolors='none', zorder=2)

# Applying the Mercator Mapping Prescription to the gridlines
grid_M = mercator(grid_dd)

# Plotting the gridlines
plt.scatter(grid_M['lambda'], grid_M['psi'], c='k', s=4, edgecolors='none', zorder=3)

# Plot title & Axis labels
plt.xlabel('$X$ $(cm)$', fontsize='16', fontweight='bold')
plt.ylabel('$Y$ $(cm)$', fontsize='16', fontweight='bold')
plt.suptitle('   Map of Locations', fontsize='26', fontweight='bold')
plt.title('(Mercator Map Prescription)', fontsize='18', style='italic')

# Adding labels to the markers
#for i, txt in enumerate(labels_M):
plt.annotate(txt, (Xm_M-8, Ym_M+4), fontsize='14', color='b')

# Tick Mark Labels
plt.xticks(np.arange(-180, 181, 30))
plt.yticks(np.arange(-80, 81, 20))
plt.ylim(-80,80)
plt.xlim(-180,180)

# Grid
plt.grid(True, color='white', zorder=1)
```

## 2.1: Coordinate Transformation Functions

**Spherical $\rightarrow$ Cartesian**

```python
def sph2xyz(l,p,h, R = 6371000):
    """
    Function that converts geocentric earth-fixed spherical coordinates 
    into geocentric earth-fixed cartesian coordinates
    
    Input
    -----
    l: lambda, longitude (units: radians)
    p: psi, latitude (units: radians)
    h: height above sphere (units: meters)
    
    Output
    ------
    x,y,z: cartesian coordinates (units: meters)
    """
    x = (R+h) * np.cos(p) * np.cos(l)
    y = (R+h) * np.cos(p) * np.sin(l)
    z = (R+h) * np.sin(p)
    
    return x,y,z
```

**Cartesian $\rightarrow$ Spherical**

```python
def xyz2sph(x,y,z, R = 6371000):
    """
    Function that converts geocentric earth-fixed cartesian coordinates into 
    geocentri earth-fixed spherical coordinates
    
    Input
    -----
    x,y,z: cartesian coordinates (units: meters)
    
    Output
    ------
    lmd: longitude (units: radians)
    psi: latitude (units: radians)
    h: height above sphere (units: meters)
    """
    lmd = np.arctan2(y, x) * deg
    psi = np.arctan2(z, math.sqrt(x**2 + y**2)) * deg
    h = math.sqrt(x**2 + y**2 + z**2) - R
    
    return lmd, psi, h
```

**Rotation Matrix**

```python
def rotate(x, y, z, omega, phi, kappa):
    """
    Function that applies rotation to geocentric earth-fixed cartesian coordinates 
    
    Input
    -----
    x,y,z: cartesian coordinates (units: meters)
    omega: angle of rotation with respect to the x-axis (units: decimal degrees)
    phi:   angle of rotation with respect to the y-axis (units: decimal degrees)
    kappa: angle of rotation with respect to the z-axis (units: decimal degrees)
    
    Output
    ------
    x,y,z: rotated cartesian coordinates (units: meters)
    """
    # Initializing Variables
    xp = np.zeros(x.shape[0])
    yp = np.zeros(y.shape[0])
    zp = np.zeros(z.shape[0])
    
    # Converting the angles from decimal degrees to radians
    ax = omega * rad
    ay = phi * rad
    az = kappa * rad
    
    # Rotation matrices for each axis
    Rx = [[1, 0, 0],
         [0, np.cos(ax), np.sin(ax)],
         [0, -np.sin(ax), np.cos(ax)]]
    
    Ry = [[np.cos(ay), 0, -np.sin(ay)],
         [0, 1, 0],
         [np.sin(ay), 0, np.cos(ay)]]
    
    Rz = [[np.cos(az), np.sin(az), 0],
         [-np.sin(az), np.cos(az), 0],
         [0, 0, 1]]
    
    # Combining the rotation matrices for each axis into a single, 3D rotation matrix
    Mxy = np.matmul(Ry,Rx)
    M = np.matmul(Rz,Mxy)
    
    # Applying rotation to cartesian coordinates (iteratively)
    for index in range(x.shape[0]):
        xp[index], yp[index], zp[index] = np.matmul(M,[x[index], y[index], z[index]])
 
    # Creating a pandas data frame of the rotated cartesian coordinates
    rotated_xyz = pd.DataFrame({
        'x': xp,
        'y': yp,
        'z': zp
    })
    
    return rotated_xyz
```

## 4.7: Measurements & Calculations

**Real World (RW)**

```python
# Square ABCD
lamA = -90 * rad
psiA = 30 * rad
lamB = -75 * rad
psiB = 30 * rad
lamC = -75 * rad
psiC = 45 * rad
lamD = -90 * rad
psiD = 45 * rad

# Distances (in meters)
ds_AB = R * np.cos(psiA) * (lamB - lamA)
ds_AC = R * np.arccos((np.sin(psiA)*np.sin(psiC)) + (np.cos(psiA)*np.cos(psiC)*np.cos(lamC - lamA)))
ds_AD = R * (psiD - psiA)
```

**Mapping World (MW) - *Plate Caree***

```python
# Square ABCD
xA_PC = -90
yA_PC = 30
xB_PC = -75
yB_PC = 30
xC_PC = -75
yC_PC = 45
xD_PC = -90
yD_PC = 45

# Distances (in centimeters)
dS_AB_PC = math.sqrt((xB_PC - xA_PC)**2 + (yB_PC - yA_PC)**2)
dS_AC_PC = math.sqrt((xC_PC - xA_PC)**2 + (yC_PC - yA_PC)**2)
dS_AD_PC = math.sqrt((xD_PC - xA_PC)**2 + (yD_PC - yA_PC)**2)
```

**Mapping World (MW) - *Sanson-Flamsteed***

```python
# Square ABCD
xA_SF = -90 * np.cos(30 * rad)
yA_SF = 30
xB_SF = -75 * np.cos(30 * rad)
yB_SF = 30
xC_SF = -75 * np.cos(45 * rad)
yC_SF = 45
xD_SF = -90 * np.cos(45 * rad)
yD_SF = 45

# Distances (in centimeters)
dS_AB_SF = math.sqrt((xB_SF - xA_SF)**2 + (yB_SF - yA_SF)**2)
dS_AC_SF = math.sqrt((xC_SF - xA_SF)**2 + (yC_SF - yA_SF)**2)
dS_AD_SF = math.sqrt((xD_SF - xA_SF)**2 + (yD_SF - yA_SF)**2)
```

**Mapping World (MW) - *Mercator***

```python
# Square ABCD
xA_M = -90
yA_M = np.log(np.tan(((30*rad)/2) + (math.pi/4))) * 180/math.pi
xB_M = -75
yB_M = np.log(np.tan(((30*rad)/2) + (math.pi/4))) * 180/math.pi
xC_M = -75
yC_M = np.log(np.tan(((45*rad)/2) + (math.pi/4))) * 180/math.pi
xD_M = -90
yD_M = np.log(np.tan(((45*rad)/2) + (math.pi/4))) * 180/math.pi

# Distances (in centimeters)
dS_AB_M = math.sqrt((xB_M - xA_M)**2 + (yB_M - yA_M)**2)
dS_AC_M = math.sqrt((xC_M - xA_M)**2 + (yC_M - yA_M)**2)
dS_AD_M = math.sqrt((xD_M - xA_M)**2 + (yD_M - yA_M)**2)
```

**Scale**

```python
# Plate Caree
sigma_AB_PC = ds_AB / ((dS_AB_PC/100))
sigma_AC_PC = ds_AC / ((dS_AC_PC/100))
sigma_AD_PC = ds_AD / ((dS_AD_PC/100))

# Sanson - Flamsteed
sigma_AB_SF = ds_AB / ((dS_AB_SF/100))
sigma_AC_SF = ds_AC / ((dS_AC_SF/100))
sigma_AD_SF = ds_AD / ((dS_AD_SF/100))

# Mercator
sigma_AB_M = ds_AB / ((dS_AB_M/100))
sigma_AC_M = ds_AC / ((dS_AC_M/100))
sigma_AD_M = ds_AD / ((dS_AD_M/100))
```

**Scales as Ratio to line AB**

```python
# Plate Caree
PC_ratio_AB_AB = sigma_AB_PC / sigma_AB_PC
PC_ratio_AB_AC = sigma_AB_PC / sigma_AC_PC
PC_ratio_AB_AD = sigma_AB_PC / sigma_AD_PC

# Sanson - Flamsteed
SF_ratio_AB_AB = sigma_AB_SF / sigma_AB_SF
SF_ratio_AB_AC = sigma_AB_SF / sigma_AC_SF
SF_ratio_AB_AD = sigma_AB_SF / sigma_AD_SF

# Mercator
M_ratio_AB_AB = sigma_AB_M / sigma_AB_M
M_ratio_AB_AC = sigma_AB_M / sigma_AC_M
M_ratio_AB_AD = sigma_AB_M / sigma_AD_M
```

**Setting Up Data Frames**

```python
# Plate Caree
pres_PC = ('Plate Caree','Plate Caree','Plate Caree')
line = ('AB','AC','AD')
RW = (ds_AB, ds_AC, ds_AD)
MW_PC = (dS_AB_PC, dS_AC_PC, dS_AD_PC)
sig_ratio_PC = ('1:' + str(int(sigma_AB_PC)), '1:' + str(int(sigma_AC_PC)), '1:' + str(int(sigma_AD_PC)))
ratio_AB_PC = (PC_ratio_AB_AB, PC_ratio_AB_AC, PC_ratio_AB_AD)

scale_PC = pd.DataFrame({
    'Pres':pres_PC,
    'Line':line, 
    'RW (m)':RW, 
    'MW (cm)':MW_PC, 
    'Scale':sig_ratio_PC,
    'Ratio to AB':ratio_AB_PC})


# Sanson-Flamsteed
pres_SF = ('Sanson-Flamsteed','Sanson-Flamsteed','Sanson-Flamsteed')
line = ('AB','AC','AD')
RW = (ds_AB, ds_AC, ds_AD)
MW_SF = (dS_AB_SF, dS_AC_SF, dS_AD_SF)
sig_ratio_SF = ('1:' + str(int(sigma_AB_SF)), '1:' + str(int(sigma_AC_SF)), '1:' + str(int(sigma_AD_SF)))
ratio_AB_SF = (SF_ratio_AB_AB, SF_ratio_AB_AC, SF_ratio_AB_AD)

scale_SF = pd.DataFrame({
    'Pres':pres_SF,
    'Line':line, 
    'RW (m)':RW, 
    'MW (cm)':MW_SF, 
    'Scale':sig_ratio_SF,
    'Ratio to AB':ratio_AB_SF})


# Mercator
pres_M = ('Mercator','Mercator','Mercator')
line = ('AB','AC','AD')
RW = (ds_AB, ds_AC, ds_AD)
MW_M = (dS_AB_M, dS_AC_M, dS_AD_M)
sig_ratio_M = ('1:' + str(int(sigma_AB_M)), '1:' + str(int(sigma_AC_M)), '1:' + str(int(sigma_AD_M)))
ratio_AB_M = (M_ratio_AB_AB, M_ratio_AB_AC, M_ratio_AB_AD)

scale_M = pd.DataFrame({
    'Pres':pres_M,
    'Line':line, 
    'RW (m)':RW, 
    'MW (cm)':MW_M, 
    'Scale':sig_ratio_M,
    'Ratio to AB':ratio_AB_M})
```

## 6.1: Conclusion

**Evaluating the Results**

As evident from the table above, scales have been calculated for all three mapping prescriptions. After evaluating them and comparing them with each other, the Plate Caree mapping prescription seems to be the best with regards to scale for the quadrant we're calling square ABCD. The second-best mapping prescription for square ABCD would be the Mercator mapping prescription, followed by the Sanson-Flamsteed. Although this order holds true for my place of birth, however, this might not be the case for other parts of the world. The best mapping prescription depends on which quadrant you are in. For example, if you live above 75 degrees latitude or below -75 degrees latitude, the Mercator mapping prescription is probably the worst in terms of scale. On the other hand, if you live in the state of Alaska or far east Asia, the Sanson-Flamsteed mapping prescription would be the worst. In contrast, however, if you live along the equator, it seems that all three of these mapping prescriptions work well for you. In that case, the Sanson-Flamsteed is still the worst. If you live along the Prime Meridian though, theSanson-Flamsteed is probably the best. In general, I would say that the Plate Caree mapping prescription is the best for most of the quadrants in the world, followed by the Mercator, then the Sanson-Flamsteed. 

As far as preserving angles, both the Plate Caree and the Mercator do a good job with this in terms of the longitude and latitude being perpendicular to each other. The Mercator mapping prescription gets worse the farther from the equator you get though. When preserving the curvature of the earth in the mapping frame, however, the Sanson-Flamsteed mapping prescription does the best job at this. 

When looking at these map scales from Lab 03 to Lab 04 to Lab 05, the behavior is inconsistent. It's almost as though we looked at two extreme examples, then looked at a happy medium. The Plate Care mapping prescription is extreme in preserving the grid of the mapping frame, so it warps the latitudes and longitudes to fit that model. On the opposite end of the spectrum, we have the Sanson-Flamsteed mapping prescription, which is extreme in preserving the angles and curvature of the earth, so it warps the grid of the mapping frame to fit the model of the earth. The Mercator mapping prescription is reminiscent of both, but doesnt go to extremes on either end. It merges the best qualities of both prescriptions, so perhaps the Mercator truly is the best mapping prescription of the three.
