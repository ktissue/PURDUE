## <span style="color:blue">*PART 2*<span>

In part 2 of this assignment, we are asked to plot the original 11 points, along with the coastline and gridline files supplied by the professor, on a map using the Plate Caree Map Prescription.

**Function for converting DMS to Decimal Degrees**

```python
def dms2dd(dd, mm, ss):
    """
    A function that will convert degree minute second (DMS) format into decimal degree format
    
    Inputs
    -----
    dd: spherical coordinate value: degrees (integer)
    mm:spherical coordinate value: minutes (integer)
    ss: spherical coordinate value: seconds (floating number)
    
    Output
    ------
    calculated decimal degree value (floating number)
    """
    return dd + mm/60 + ss/3600
```

**Initializing the Arrays**

```python
lmd_dd = np.zeros(11, dtype = np.float64)
psi_dd = np.zeros(11, dtype = np.float64)
```

**Applying the Conversion from DMS to Decimal Degrees**

```python
for index, row in sph_dms.iterrows():
    lmd_dd[index] = dms2dd(row['$\lambda$dd'], row['$\lambda$mm'], row['$\lambda$ss.sssss'])
    psi_dd[index] = dms2dd(row['$\psi$dd'], row['$\psi$mm'], row['$\psi$ss.sssss'])
```

**Reading in the coastline file**

```python
coastline = pd.read_csv('coastline.dat', header=None, sep='\s\s+', engine='python', usecols=[1,2], names=["psi","lambda"])
coast_dd = coastline/60
```

**Reading in the gridline file**

```python
gridline = pd.read_csv('gridline.dat', header=None, sep='\s\s+', engine='python', usecols=[1,2], names=["psi","lambda"])
grid_dd = gridline/60
```

## 2.1: Plotting the Map

**Plate Caree Mapping Prescription**

```python
Xm = lmd_dd
Ym = psi_dd
```

```python
# Labels for markers
labels = ['  NP', '  SP', '  PoB', '  1', '   2', '  3', '  4', '    5', '  6', '  7', '    8']

# Resizing the plot window
plt.rcParams["figure.figsize"] = (14,7)

# Plotting the original locations
plt.plot(Xm, Ym, 'ro', markersize=8, zorder=4)

# Plotting the coastlines
plt.scatter(coast_dd['lambda'], coast_dd['psi'], c='k', s=4, edgecolors='none', zorder=2)

# Plotting the gridlines
plt.scatter(grid_dd['lambda'], grid_dd['psi'], c='k', s=4, edgecolors='none', zorder=3)

# Plot title & Axis labels
plt.xlabel('$\lambda °$', fontsize='16', fontweight='bold')
plt.ylabel('$\psi °$', fontsize='16', fontweight='bold')
plt.suptitle('   Map of Locations', fontsize='26', fontweight='bold')
plt.title('(Plate Caree Map Prescription)', fontsize='18', style='italic')

# Adding labels to the markers
for i, txt in enumerate(labels):
    plt.annotate(txt, (Xm[i], Ym[i]), fontsize='14', color='b')

# Tick Mark Labels
plt.xticks(np.arange(-180, 181, 15), fontsize='14', rotation=45)
plt.yticks(np.arange(-90, 91, 15), fontsize='14')

# Grid
plt.grid(True, color='white', zorder=1)
```

## 3.2: Measurements & Calculations

**Real World (RW)**

```python
lamA = -90 * rad
psiA = 30 * rad
lamB = -75 * rad
psiB = 30 * rad
lamC = -75 * rad
psiC = 45 * rad
lamD = -90 * rad
psiD = 45 * rad
```

**RW distances *(in meters)***

```python
ds_AB = R * np.cos(psiA) * (lamB - lamA)
ds_AC = R * np.arccos((np.sin(psiA)*np.sin(psiC)) + (np.cos(psiA)*np.cos(psiC)*np.cos(lamC - lamA)))
ds_AD = R * (psiD - psiA)
```

**Mapping World (MW)**

```python
xA = -90
yA = 30
xB = -75
yB = 30
xC = -75
yC = 45
xD = -90
yD = 45
```

**MW distances *(in centimeters)***

```python
dS_AB = math.sqrt((xB - xA)**2 + (yB - yA)**2)
dS_AC = math.sqrt((xC - xA)**2 + (yC - yA)**2)
dS_AD = math.sqrt((xD - xA)**2 + (yD - yA)**2)
```

**Scale**

```python
sigma_AB = 1 / ((dS_AB/100) / ds_AB)
sigma_AC = 1 / ((dS_AC/100) / ds_AC)
sigma_AD = 1 / ((dS_AD/100) / ds_AD)
```

**Setting up a data frame**

```python
line = ('AB','AC','AD')
RW = (ds_AB, ds_AC, ds_AD)
MW = (dS_AB, dS_AC, dS_AD)
sig_ratio = ('1:' + str(int(sigma_AB)), '1:' + str(int(sigma_AC)), '1:' + str(int(sigma_AD)))

scale = pd.DataFrame({
    'Line':line, 
    'RW (m)':RW, 
    'MW (cm)':MW, 
    'Scale':sig_ratio
})
```

**Displaying Scale in Table**

```python
print('Scale')
print()
print("-------------------------------------------------------------")
print("  Line        RW dist(m)      MW dist(cm)           Scale")
print("-------------------------------------------------------------")
for index, row in scale.iterrows():
    print("%5s %18.3f %13.3f %20s" % (row['Line'], row['RW (m)'], row['MW (cm)'], row['Scale']))
print("-------------------------------------------------------------")
```

## 4.1: Conclusion

The scale of this map is different depending on where in the map you are located, which would make it very challenging to determine distances on the fly without the use of a computer.  Lines in the latitudinal direction, however, such as line AD, maintain a uniform scale due to all of the lines in that direction being great circles. That scale is a function of the change in latitude, which for us was given as 15 degrees. Since those lines converge at the poles, though, the distances in the longitudinal directions will be different depending upon which latitude you are located. The closer you are to the poles, the larger the scale will be. Likewise, the closer you are to the equator, the smaller the scale will be. For square ABCD, the longitude ranges from -90 degrees to -75 degrees and the latitude ranges from 30 degrees to 45 degrees. This means that the scale at the bottom of the map will be smaller than the scale at the top of the map. When traveling diagonally, we need to incorporate the law of cosines to account for the surface distance in the real world. Overall, these scales make sense in terms of their relationship to each other and the map itself is visually pleasing, however, the amount of work involved to plot information accurately using the Plate Caree map prescription makes its day-to-day utility a little difficult to justify.
