### Importing required module

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Plot figures inside the notebook
%matplotlib inline
```

Then initializing variables

- $\rho \rightarrow$ conversion factor from radian to degree
- $\alpha_{1} \rightarrow$ clockwise angle between $\frac{3 \pi}{2}$ and $\pi$ with 5 degree interval
- $\alpha_{2} \rightarrow$ clockwise angle between 0 and $\frac{3 \pi}{2}$ with 5 degree interval
- $\alpha_{3} \rightarrow$ clockwise angle between $\frac{\pi}{2}$ and $2\pi$ with 5 degree interval
- $\alpha_{4} \rightarrow$ clockwise angle between $\pi$ and $\frac{\pi}{2}$ with 5 degree interval
- $R \rightarrow$ radius of the Earth (approximately 6371 km)

**Note:** Your markdown above and your code below is not matching, especially for $\alpha_1$. This made me spent quite some time to figure out why your program works! 

```python
rho = 180 / math.pi
alpha1 = np.arange((3/2) * math.pi, 3 * math.pi+0.1, math.pi/36.0)
alpha2 = np.arange(0, (3/2) * math.pi+0.1, math.pi/36.0)
alpha3 = np.arange((1/2) * math.pi, 2 * math.pi+0.1, math.pi/36.0)
alpha4 = np.arange(math.pi, (5/2) * math.pi+0.1, math.pi/36.0)
R = 6371
```

```python
alpha1*rho
```

```python
alpha2*rho
```

Now convert polar coordinates into cartesian coordinates using below equations

- $x_{1} = R \cdot cos(\alpha_{1})$
- $y_{1} = R \cdot sin(\alpha_{1})$
- $x_{2} = R \cdot cos(\alpha_{2})$
- $y_{2} = R \cdot sin(\alpha_{2})$
- $x_{3} = R \cdot cos(\alpha_{3})$
- $y_{3} = R \cdot sin(\alpha_{3})$
- $x_{4} = R \cdot cos(\alpha_{4})$
- $y_{4} = R \cdot sin(\alpha_{4})$

```python
x1 = R * np.cos(alpha1) + R
y1 = R * np.sin(alpha1) + R
x2 = R * np.cos(alpha2) - R
y2 = R * np.sin(alpha2) + R
x3 = R * np.cos(alpha3) - R
y3 = R * np.sin(alpha3) - R
x4 = R * np.cos(alpha4) + R
y4 = R * np.sin(alpha4) - R
```

Converting alpha to degrees

```python
alpha_deg1 = alpha1 * rho
alpha_deg2 = alpha2 * rho
alpha_deg3 = alpha3 * rho
alpha_deg4 = alpha4 * rho
```

```python
alpha_deg1
```

Removing 360 degrees where necessary (for display purposes)

```python
for i in range(0 - len(alpha_deg1)):
    if alpha_deg1[i] >= 360:
        alpha_deg1[i] = alpha_deg1[i] - 360
```

Now making a data frame for each quadrant

```python
results1 = pd.DataFrame({'alpha_deg': alpha_deg1, 'x': x1, 'y': y1}) # Quadrant 1
results2 = pd.DataFrame({'alpha_deg': alpha_deg2, 'x': x2, 'y': y2}) # Quadrant 2
results3 = pd.DataFrame({'alpha_deg': alpha_deg3, 'x': x3, 'y': y3}) # Quadrant 3
results4 = pd.DataFrame({'alpha_deg': alpha_deg4, 'x': x4, 'y': y4}) # Quadrant 4
```

```python
results1
```

Concatenating the results into a single data frame

```python
results = results1.append(results2)
results = results.append(results3)
results = results.append(results4)
```

Printing the data frame with 3 decimal points

```python
pd.options.display.float_format = '{:,.3f}'.format
print(results)
```

Now let's plot the graph

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x1,y1, 'go') # Plotting the clover leaf in quadrant 1
ax.plot(x2,y2, 'go') # Plotting the clover leaf in quadrant 2
ax.plot(x3,y3, 'go') # Plotting the clover leaf in quadrant 3
ax.plot(x4,y4, 'go') # Plotting the clover leaf in quadrant 4
ax.plot(-R,R, 'r*') # Plotting the center of the clover leaf in quadrant 1
ax.plot(R,R, 'r*') # Plotting the center of the clover leaf in quadrant 2
ax.plot(R,-R, 'r*') # Plotting the center of the clover leaf in quadrant 3
ax.plot(-R,-R, 'r*') # Plotting the center of the clover leaf in quadrant 4
ax.axis('equal') # axis
ax.grid(True, alpha=0.4, linestyle='--') # grid
ax.set_xlabel('X (km)', fontsize=12, fontweight='bold') # x-axis label
ax.set_ylabel('Y (km)', fontsize=12, fontweight='bold') # y-axis label
ax.set_title('CE 597 Homework #3 - Kevan Tissue', fontsize=16, fontweight="bold") # title
```
